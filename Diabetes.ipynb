```python
def classify_bmi(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 25:
        return 'Normal'
    elif 25 <= bmi < 30:
        return 'Overweight'
    else:
        return 'Obese'

df_5000['BMI_Category'] = df_5000['BMI'].apply(classify_bmi)

print(df_5000[['BMI', 'BMI_Category']].head())

print(df_5000['BMI_Category'].value_counts())

sns.countplot(x='BMI_Category', data=df_5000, order=['Underweight', 'Normal', 'Overweight', 'Obese'])
plt.title('BMI Category Distribution')
plt.show()
```

```python
pd.crosstab(df_5000['Smoking_Status'], df_5000['BMI_Category'], normalize='index')
```

```python
def classify_waist(waist):
    if waist < 80:
        return 'Low Risk'
    elif 80 <= waist < 94:
        return 'Moderate Risk'
    else:
        return 'High Risk'

df_5000['Waist_Category'] = df_5000['Waist_Circumference'].apply(classify_waist)
```

```python
def classify_fbg(fbg):
    if fbg < 100:
        return 'Normal'
    elif 100 <= fbg < 126:
        return 'Prediabetes'
    else:
        return 'Diabetes'

df_5000['FBG_Category'] = df_5000['Fasting_Blood_Glucose'].apply(classify_fbg)
```

```python
def classify_hba1c(hba1c):
    if hba1c < 5.7:
        return 'Normal'
    elif 5.7 <= hba1c < 6.5:
        return 'Prediabetes'
    else:
        return 'Diabetes'

df_5000['HbA1c_Category'] = df_5000['HbA1c'].apply(classify_hba1c)
```

```python
pd.crosstab(df_5000['HbA1c_Category'], df_5000['Physical_Activity_Level'])
```

```python
def classify_systolic(bp):
    if bp < 120:
        return 'Normal'
    elif 120 <= bp < 130:
        return 'Elevated'
    elif 130 <= bp < 140:
        return 'Hypertension Stage 1'
    else:
        return 'Hypertension Stage 2'

df_5000['BP_Systolic_Category'] = df_5000['Blood_Pressure_Systolic'].apply(classify_systolic)

pd.crosstab(df_5000['Smoking_Status'], df_5000['BP_Systolic_Category'])
```

```python
def classify_diastolic(bp):
    if bp < 80:
        return 'Normal'
    elif 80 <= bp < 90:
        return 'Hypertension Stage 1'
    else:
        return 'Hypertension Stage 2'

df_5000['BP_Diastolic_Category'] = df_5000['Blood_Pressure_Diastolic'].apply(classify_diastolic)

pd.crosstab(df_5000['Smoking_Status'], df_5000['BP_Diastolic_Category'])
```

```python
def classify_total_chol(chol):
    if chol < 200:
        return 'Desirable'
    elif 200 <= chol < 240:
        return 'Borderline High'
    else:
        return 'High'

df_5000['Chol_Total_Category'] = df_5000['Cholesterol_Total'].apply(classify_total_chol)

pd.crosstab(df_5000['Smoking_Status'], df_5000['Chol_Total_Category'])
```

```python
def classify_hdl(hdl):
    if hdl < 40:
        return 'Low'
    elif hdl >= 60:
        return 'High'
    else:
        return 'Acceptable'

df_5000['HDL_Category'] = df_5000['Cholesterol_HDL'].apply(classify_hdl)

pd.crosstab(df_5000['Smoking_Status'], df_5000['HDL_Category'])
```

```python
def classify_ldl(ldl):
    if ldl < 100:
        return 'Optimal'
    elif 100 <= ldl < 130:
        return 'Near Optimal'
    elif 130 <= ldl < 160:
        return 'Borderline High'
    elif 160 <= ldl < 190:
        return 'High'
    else:
        return 'Very High'

df_5000['LDL_Category'] = df_5000['Cholesterol_LDL'].apply(classify_ldl)

pd.crosstab(df_5000['Smoking_Status'], df_5000['LDL_Category'])
```

```python
def classify_ggt(ggt):
    if ggt < 40:
        return 'Normal'
    elif 40 <= ggt < 80:
        return 'Moderate'
    else:
        return 'High'

df_5000['GGT_Category'] = df_5000['GGT'].apply(classify_ggt)

pd.crosstab(df_5000['Smoking_Status'], df_5000['GGT_Category'])
```

```python
def classify_urate(urate):
    if urate < 4:
        return 'Low'
    elif 4 <= urate <= 7:
        return 'Normal'
    else:
        return 'High'

df_5000['Urate_Category'] = df_5000['Serum_Urate'].apply(classify_urate)

pd.crosstab(df_5000['Smoking_Status'], df_5000['Urate_Category'])
```

```python
def classify_calories(cal):
    if cal < 1500:
        return 'Low Intake'
    elif 1500 <= cal < 2500:
        return 'Normal Intake'
    else:
        return 'High Intake'

df_5000['Calories_Category'] = df_5000['Dietary_Intake_Calories'].apply(classify_calories)

pd.crosstab(df_5000['Smoking_Status'], df_5000['Calories_Category'])
```
